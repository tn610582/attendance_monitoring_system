# Haar Face Detection:

 - Task 1: Read input file.
 - Task 2: Use harr algorithm detect face.
 - Task 3: Draw bounding box.

```python
# Imports:
import cv2
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
plt.rcParams['figure.dpi'] = 150

# Load image and plot
img = cv2.imread('./images/fake_ai_faces.0.png')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

# Create haar detector we use front face detector here
detector = cv2.CascadeClassifier('./haarcascades/haarcascade_frontalface_default.xml')

# Convert image to grayscale image
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

# Detect face
# scaleFactor -> size of image
# minNeighbors -> number of propose bbox
# minSize -> bbox size filter
# maxSize -> bbox size filter
detection = detector.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=10)

# Show detection
#print(detection)

# Draw bbox on image
for (x,y,w,h) in detection:
    #draw bbox on img
    cv2.rectangle(img, (x,y), (x+w,y+h),(0,0,255),5)

# Show bbox on image
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
