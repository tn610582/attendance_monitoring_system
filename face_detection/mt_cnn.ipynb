# MT CNN Face Detection:

 - Task 1: Read input file.
 - Task 2: Use cnn algorithm detect face.
 - Task 3: Draw bounding box.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline
plt.rcParams['figure.dpi'] = 150
img = cv2.imread('./images/fake_ai_faces.0.png')
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
from mtcnn.mtcnn import MTCNN
```

```python
# Create detector
face_detection = MTCNN()
```

```python
detections = face_detection.detect_faces(img)
for face in detections:
    (x,y,w,h) = face['box']
    cv2.rectangle(img,(x,y),(x+w,y+w),(0,255,0),5)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
