# CNN Face Detection:

 - Task 1: Read input file.
 - Task 2: Use harr algorithm detect face.
 - Task 3: Draw bounding box.

```python
# Import Dlib
import dlib
import cv2
import matplotlib.pyplot as plt
#%matplotlib inline
plt.rcParams['figure.dpi'] = 150
img = cv2.imread('./images/fake_ai_faces.0.png')
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
# Create detector
face_detection = dlib.cnn_face_detection_model_v1('./weights/mmod_human_face_detector.dat')
```

```python
detection = face_detection(img,1)
```

```python
# decode detection data
for face in detection:
    x = face.rect.left()
    y = face.rect.top()
    r = face.rect.right()
    b = face.rect.bottom()
    # confidence score
    c = face.confidence
    #print(c)
    cv2.rectangle(img, (x,y), (r,b), (0,0,255),5)
```
